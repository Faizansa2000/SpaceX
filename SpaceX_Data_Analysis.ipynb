```python
success_rate = (spacex['Mission Outcome'].value_counts(normalize=True))*100

plt.figure(figsize = (8,4))
ax1= sns.countplot(x = 'Vehicle Type', hue = 'Mission Outcome', data = spacex, palette = 'Blues')
legend_labels,_ = ax1. get_legend_handles_labels ()
ax1. legend (bbox_to_anchor=(1,1))
plt.title('Mission Outcomes', size = 20)
plt.xlabel('Vehicle Types')
plt.ylabel('Mission Count')
plt. legend (['Failure', 'Success'])
```

```python
missions_per_year = spacex.groupby([spacex['Launch Date'].dt.year, 'Mission Outcome'])['Payload Name'].count().unstack(fill_value=0)
missions_per_year['Success Rate'] = missions_per_year['Success']
missions_per_year['Failure Rate'] = missions_per_year['Failure']
plt.figure(figsize=(10, 6))
plt.plot(missions_per_year.index, missions_per_year['Success Rate'], marker='o', linestyle='-', label='Success Rate')
plt.plot(missions_per_year.index, missions_per_year['Failure Rate'], marker='o', linestyle='-', label='Failure Rate')
plt.title('Success and Failure Over the Years')
plt.xlabel('Year')
plt.ylabel('Missions')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize = (12,6))
ax1= sns.countplot(x = 'Launch Site', hue = 'Mission Outcome', data = spacex, palette = 'Greens')
legend_labels,_ = ax1. get_legend_handles_labels ()
ax1. legend (bbox_to_anchor=(1,1))
plt.title('Mission Outcomes', size = 20)
plt.xlabel('Launch Site')
plt.ylabel('Mission Count')
plt. legend (['Failure', 'Success'])
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

```python
country_counts = spacex['Customer Country'].value_counts().sort_values(ascending=False)
colors = plt.cm.tab20.colors
explode = [0.08 if i == 0 else 0.05 for i in range(len(country_counts))]
plt.figure(figsize=(8, 8))
plt.title('Percentage of Missions per Country')
plt.pie(country_counts, autopct='%.2f%%', labels=country_counts.index, colors=colors, explode=explode)
plt.show()
```

```python
launches_per_year = spacex.groupby(spacex['Launch Date'].dt.year)['Payload Name'].count()
plt.figure(figsize=(10, 6))
plt.plot(launches_per_year.index, launches_per_year.values, marker='o', linestyle='-')
plt.title('Number of Launches Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Launches')
plt.grid(True)
plt.show()
```
